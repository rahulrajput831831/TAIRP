# Comparative Analysis: VADER vs RoBERTa

## Objective

The objective of this notebook is to compare sentiment predictions generated by:

- VADER Sentiment Analyzer
- RoBERTa Transformer Model

Comparison is performed using:
- sentiment distribution
- prediction behavior
- computational efficiency
- contextual understanding

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

plt.style.use('ggplot')

print("Libraries Imported Successfully 🚀")
```

```python
vader = pd.read_csv(
    "../data/vader_sentiment_results.csv"
)

roberta = pd.read_csv(
    "../data/roberta_sentiment_results.csv"
)

print("Files Loaded Successfully ✅")
```

```python
vader.head()
```

```python
roberta.head()
```

```python
vader_counts = (
    vader['sentiment']
    .value_counts()
)

roberta_counts = (
    roberta['roberta_sentiment']
    .value_counts()
)

print(vader_counts)

print()

print(roberta_counts)
```

```python
comparison = pd.DataFrame({

'VADER': vader_counts,

'RoBERTa': roberta_counts

}).fillna(0)

comparison
```

```python
comparison.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title(
    "VADER vs RoBERTa Comparison"
)

plt.xlabel(
    "Sentiment"
)

plt.ylabel(
    "Tweet Count"
)

plt.show()
```

```python
fig, axes = plt.subplots(
    1,
    2,
    figsize=(14,7)
)

vader_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    ax=axes[0]
)

axes[0].set_title(
    "VADER"
)

roberta_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    ax=axes[1]
)

axes[1].set_title(
    "RoBERTa"
)

plt.show()
```

```python
performance = pd.DataFrame({

'Metric':[

'Architecture',
'Speed',
'Context',
'Accuracy'

],

'VADER':[

'Rule-Based',
'Fast',
'Limited',
'Moderate'

],

'RoBERTa':[

'Transformer',
'Slower',
'Strong',
'Higher'

]

})

performance
```

# Final Conclusion

Findings:

- VADER produced faster predictions.
- RoBERTa captured contextual meaning more effectively.
- RoBERTa showed stronger sentiment understanding.
- VADER remained computationally efficient.

Final Recommendation:

For higher quality sentiment analysis:
→ RoBERTa

For faster execution:
→ VADER
